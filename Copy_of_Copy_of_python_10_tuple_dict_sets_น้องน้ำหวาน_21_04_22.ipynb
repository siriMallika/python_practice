```python
def reverse(d):
```

```python
def reverse(d):
  lang = {}
  lang_list=[]
  p=set()
  for k,v in d.items():
    lang_list.append(v)  # computer programming language  lang_list
    p.add(k)       # programmers' names  p 
    lang[v]={} # lang dict key  value
  print(lang_list)  # ['Python', 'C', 'C']
  print(p) #  {'Kob', 'Pear', 'Tom'}
  for i in lang_list: 
    inp=set()
    for j in p:
      if i==d[j]:  
        inp.add(j)
    lang[i]=inp
  print(lang)
```
